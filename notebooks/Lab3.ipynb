##### Question 1

This question is straightforward, and only focuses on sampling from a multivariate normal distribution with different covariance matrices. The idea of the exercise is to notice that the covariance matrix defines the dispersion of points in the sample. The diagaonal entries in the covariance matrix defines the dispersion (variance) along the respective axis. The non-diagonal elements govern the dispersion between the axis, if the first axis is fixed. Observe the difference in variance between Sample 2 and Sample 3. Notice that Sample 3 appears to be less random. This is becuase the covariance between the two variables is higher, so once the first point is fixed, the second is likely to be in the same neighborhood. 

```python
np.random.seed(115)
```

```python
Cov_1 = np.array([[1, 0], [0 ,2]])
Cov_2 = np.array([[5, 1], [1, 2]])
Cov_3 = np.array([[5, 2], [2, 2]])
Cov_4 = np.array([[5, 2], [2, 1]])

mean = np.zeros(2)
```

```python
sample1 = np.random.multivariate_normal(mean, Cov_1, size=200)
sample2 = np.random.multivariate_normal(mean, Cov_2, size=200)
sample3 = np.random.multivariate_normal(mean, Cov_3, size=200)
sample4 = np.random.multivariate_normal(mean, Cov_4, size=200)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid of subplots
axes = axes.flatten() 

samples = [sample1, sample2, sample3, sample4]

for i, sample in enumerate(samples):
    ax = axes[i]
    ax.scatter(sample[:, 0], sample[:, 1], alpha=0.7, color='red', marker='*')
    ax.set_title(f"Sample {i+1}")
    ax.set_xlabel("X")
    ax.set_ylabel("X")
    ax.grid(True)

plt.tight_layout()
plt.show()
```
